```python
import os
import numpy as np
from PIL import Image
import tensorflow as tf
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten,BatchNormalization
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras import backend as K
from tensorflow.keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt
import cv2
import pickle
```

```python
pickle_in = open("Xd.pickle","rb")
X = pickle.load(pickle_in)


pickle_in = open("yd.pickle","rb")
y = pickle.load(pickle_in)

X = X/255.0

X=np.array(X)
y=np.asarray(y)

model = Sequential()

#4 layers
model.add(Conv2D(input_shape=(100,100,1),filters = 128,kernel_size=(3,3),strides = (1,1),padding='same',activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(input_shape=(100,100,1),filters = 128,kernel_size=(3,3),strides = (1,1),padding='same',activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(input_shape=(100,100,1),filters = 256,kernel_size=(3,3),strides = (1,1),padding='same',activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(input_shape=(100,100,1),filters = 256,kernel_size=(3,3),strides = (1,1),padding='same',activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))

#model.add(Dropout(0.25))

model.add(Flatten())  # this converts our 3D feature maps to 1D feature vectors

model.add(Dense(128,activation = 'relu'))
model.add(BatchNormalization())
model.add(Dense(64,activation = 'relu'))
model.add(BatchNormalization())
model.add(Dense(10,activation='softmax'))

model.add(Activation('softmax'))

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(X, y, batch_size=32, epochs=10, validation_split=0.25)

model.save('CNNDistanceTransform1.model')
model.save_weights('CNNDistanceTransform1WEIGHTS.h5')
```
